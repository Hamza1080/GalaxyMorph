# Terms identifying

| Task  | Question                                      | Example Answers                                      |
|-------|-----------------------------------------------|------------------------------------------------------|
| t01   | Is the galaxy smooth or does it have features? | a01: smooth, a02: features/disk, a03: star/artifact  |
| t02   | Is the galaxy edge-on? (if features/disk)      | a04: yes, a05: no                                    |
| t03   | Does the galaxy have a bar?                    | a06: bar, a07: no bar                                |
| t04   | Does the galaxy have spiral arms?              | a08: spiral, a09: no spiral                          |
| t05   | Bulge prominence?                              | a10–a13: no, just noticeable, obvious, dominant      |
| t06   | Is anything odd?                               | a14: yes, a15: no                                    |
| t07   | How rounded is the galaxy?                     | a16: completely round, a17: in-between, a18: cigar   |
| t08   | What odd features are present?                 | a19–a24, a38: rings, lens/arc, merger, dust lane, etc. |
| t09   | Bulge shape?                                   | a25: rounded, a26: boxy, a27: no bulge               |
| t10   | Spiral arm winding?                            | a28–a30: tight, medium, loose                        |
| t11   | Number of spiral arms?                         | a31–a37: 1, 2, 3, 4, >4, can't tell                  |

```python
ellipticals = newdf[(newdf['to1_a01'] > 0.8) & (newdf['to1_a02'] < 0.2)]
ellipticals
```

```python
# Classifying elliptical galaxies into E0-E1, E2-E5, E6-E7 based on roundness
newdf.loc[
    (newdf['to1_a01'] > 0.8) & (newdf['to1_a02'] < 0.2) & (newdf['to7_a16'] > 0.6),
    'class'
] = 'E0-E1'

newdf.loc[
    (newdf['to1_a01'] > 0.8) & (newdf['to1_a02'] < 0.2) & (newdf['to7_a17'] > 0.6),
    'class'
] = 'E2-E5'

newdf.loc[
    (newdf['to1_a01'] > 0.8) & (newdf['to1_a02'] < 0.2) & (newdf['to7_a18'] > 0.5),
    'class'
] = 'E6-E7'

# Optionally, you can check the distribution after these updates
print(newdf['class'].value_counts())
```

```python

newdf.loc[
    (newdf['to1_a02'] > 0.8) & (newdf['to2_a04'] > 0.7) & (newdf['to3_a06'] > 0.5),
    'class'
] = 'Edge-on Barred Spiral'

newdf.loc[
    (newdf['to1_a02'] > 0.8) & (newdf['to2_a04'] > 0.7) & (newdf['to3_a07'] > 0.5),
    'class'
] = 'Edge-on Unbarred Spiral'

newdf.loc[
    (newdf['to1_a02'] > 0.5) & (newdf['to2_a05'] > 0.7) & (newdf['to4_a08'] > 0.6) & (newdf['to3_a06'] > 0.5),
    'class'
] = 'Barred Spiral'

newdf.loc[
    (newdf['to1_a02'] > 0.5) & (newdf['to2_a05'] > 0.7) & (newdf['to4_a08'] > 0.6) & (newdf['to3_a07'] > 0.5),
    'class'
] = 'Unbarred Spiral'

spiral_classes = newdf.loc[newdf['class'].notnull(), ['dr7objid', 'sample', 'gz2_class', 'class']].copy()
spiral_classes
```

```python
# Start by creating a copy of newdf
irregular_df = newdf.copy()

# Apply classification rules in order of specificity
irregular_df.loc[
    (irregular_df['to8_a19'] > 0.3) & (irregular_df['to6_a14'] > 0.5),
    'class'
] = 'ring'

irregular_df.loc[
    (irregular_df['to8_a23'] > 0.3) & (irregular_df['to6_a14'] > 0.5) &
    (irregular_df['class'].isnull()),
    'class'
] = 'peculiar'

irregular_df.loc[
    (irregular_df['to6_a14'] > 0.5) &
    (irregular_df['class'].isnull()),
    'class'
] = 'irregular'

# Optional: pick only relevant columns if you want to isolate results
irregular_classes = irregular_df.loc[
    irregular_df['class'].notnull(),
    ['dr7objid', 'sample', 'gz2_class', 'class']
].copy()

# Check class distribution
print(irregular_classes['class'].value_counts())
```

```python
import pandas as pd
from PIL import Image, ImageEnhance
import os
import random
import uuid

# Paths
image_folder = 'images'            # Folder containing original images
output_folder = 'dip'              # Folder to save processed images
os.makedirs(output_folder, exist_ok=True)

# Load your DataFrame (assuming you already have one)
df = dataframe.copy()              # Replace with your actual DataFrame if needed

# List to store metadata of saved images
records = []

# Loop through the DataFrame
count=0
for _, row in df.iterrows():
    original_asset_id = row['asset_id']
    class_label = row['class']
    filename = f"{original_asset_id}.jpg"
    image_path = os.path.join(image_folder, filename)

    if not os.path.exists(image_path):
        print(f"Image not found: {filename}")
        continue

#     # Load original image
    img = Image.open(image_path)
    img = img.resize((256, 256))

    # Apply DIP if class matches
    if class_label == 'Edge-on Barred Spiral' or class_label=="E6-E7":
        count=count+1
        new_filename = f"{count}.jpg"
        original_save_path = os.path.join(output_folder, new_filename)
        img.save(original_save_path)
        records.append({'Asset_id': count, 'class': class_label})
        
        min_value = 1.1
        max_value = 2.4

        valid_multiples = [x for x in range(0, 361, 15)]
        num_augmentations = random.randint(6, 7)  # Random number of augmentations per image

        random_rotations = random.sample(valid_multiples, num_augmentations)
        brightness_values = [round(random.uniform(min_value, max_value), 2) for _ in range(num_augmentations)]

        for i in range(num_augmentations):
            count=count+1
            print(count)
            transformed = img.rotate(random_rotations[i])
            transformed = ImageEnhance.Brightness(transformed).enhance(brightness_values[i])

#             # Generate a new unique asset ID for the augmented image
            new_filename = f"{count}.jpg"
            output_path = os.path.join(output_folder, new_filename)
            transformed.save(output_path)

#             # Save record
            records.append({'Asset_id': count, 'class': class_label})
    elif class_label == "ring" or class_label=="Edge-on Unbarred Spiral" or class_label=="peculiar" or class_label=="irregular":
        count=count+1
        new_filename = f"{count}.jpg"
        original_save_path = os.path.join(output_folder, new_filename)
        img.save(original_save_path)
        records.append({'Asset_id': count, 'class': class_label})
        
        min_value = 1.1
        max_value = 2.4

        valid_multiples = [x for x in range(0, 361, 15)]
        num_augmentations = random.randint(2, 3)  # Random number of augmentations per image

        random_rotations = random.sample(valid_multiples, num_augmentations)
        brightness_values = [round(random.uniform(min_value, max_value), 2) for _ in range(num_augmentations)]

        for i in range(num_augmentations):
            count=count+1
            print(count)
            transformed = img.rotate(random_rotations[i])
            transformed = ImageEnhance.Brightness(transformed).enhance(brightness_values[i])

#             # Generate a new unique asset ID for the augmented image
            new_filename = f"{count}.jpg"
            output_path = os.path.join(output_folder, new_filename)
            transformed.save(output_path)

#             # Save record
            records.append({'Asset_id': count, 'class': class_label})
    else:
        count=count+1
        new_filename = f"{count}.jpg"
        transformed = ImageEnhance.Brightness(img).enhance(1.2)
        output_path = os.path.join(output_folder, new_filename)
        transformed.save(output_path)
        records.append({'Asset_id': count, 'class': class_label})
        
# # Save metadata of saved images
df_augmented = pd.DataFrame(records)
df_augmented.to_csv('processed_image_info.csv', index=False)

print("Processing complete. Info saved in 'processed_image_info.csv'.")
```
